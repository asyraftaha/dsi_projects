## Train DataFrame Cleaning

**Assigning Ordinal Values Ranking**

```python
train[['Paved Drive']] = train[['Paved Drive']].replace(['N','P','Y'],[0,1,2])

train[['Garage Finish']] = train[['Garage Finish']].replace([np.nan,'Unf','RFn','Fin'],[0,1,2,3])

train[['Alley']] = train[['Alley']].replace(['Grvl','Pave'],[1,2])

train[['Street']] = train[['Street']].replace(['Grvl','Pave'],[1,2])

train[['Land Slope']] = train[['Land Slope']].replace(['Sev','Mod','Gtl'],[1,2,3])

train[['Central Air']] = train[['Central Air']].replace(['N','Y'],[0,1])

train[['Functional']] = train[['Functional']].replace(['Sal','Sev','Maj2','Maj1','Mod','Min2','Min1','Typ'],[1,2,3,4,5,6,7,8])

train[['Fence']] = train[['Fence']].replace([np.nan,'MnWw','GdWo','MnPrv','GdPrv'],[0,1,2,3,4])

train[['BsmtFin Type 1','BsmtFin Type 2']] = train[['BsmtFin Type 1','BsmtFin Type 2']].replace([np.nan,'Unf','LwQ','Rec','BLQ','ALQ','GLQ'],[0,1,2,3,4,5,6])

train[['Bsmt Exposure']] = train[['Bsmt Exposure']].replace([np.nan,'No','Mn','Av','Gd'],[0,1,2,3,4])

train[['Exter Qual','Exter Cond','Bsmt Qual',
      'Bsmt Cond','Heating QC','Kitchen Qual',
      'Fireplace Qu','Garage Qual','Garage Cond',
     'Pool QC']] = train[['Exter Qual','Exter Cond','Bsmt Qual',
      'Bsmt Cond','Heating QC','Kitchen Qual',
      'Fireplace Qu','Garage Qual','Garage Cond',
     'Pool QC']].replace([np.nan,'Po','Fa','TA','Gd','Ex'],[0,1,2,3,4,5])
```

**Dealing with Missing Values**

```python
#Visualising the missing values
plt.figure(figsize = (32,20))
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis');
```

```python
#Assigning variable to null values columns list
null_columns = train.columns[train.isnull().sum()>0].tolist()
```

```python
#Checking list
train[null_columns].isnull().sum()
```

```python
#Pairing Mas Vnr Type and Mas Vnr Area null values
train[train['Mas Vnr Type'].isnull()][['Mas Vnr Type','Mas Vnr Area']]
```

```python
#Filling null values
train['Mas Vnr Type'].fillna('None',inplace=True)
train['Mas Vnr Area'].fillna(0,inplace=True)
```

Since the Mas Vnr Type and Mas Vnr Area have the same null values, we can assume that there is indeed no Mas Vnr present for the house and can fill the values with None and 0.

```python
train[train['Total Bsmt SF'].isnull()]
```

```python
train['Total Bsmt SF'].fillna(0,inplace=True)
train['BsmtFin SF 1'].fillna(0,inplace=True)
train['BsmtFin SF 2'].fillna(0,inplace=True)
train['Bsmt Unf SF'].fillna(0,inplace=True)
```

For these columns ['Total Bsmt SF','BsmtFin SF 1','BsmtFin SF 2','Bsmt Unf SF'], since there is only one null value and it coincides together in accordance to other null basement columns, we can assume that there is no basement for this observation and can fill the missing value with 0.

```python
train[train['Bsmt Full Bath'].isnull()]
```

```python
train['Bsmt Full Bath'].fillna(0,inplace=True)
train['Bsmt Half Bath'].fillna(0,inplace=True)
```

Again, from the two observations narrowed above, it is clear that for these observations, there is no basement for the house and can be filled with 0

```python
train[train['Garage Cars'].isnull()]
```

```python
train['Garage Cars'].fillna(0,inplace=True)
train['Garage Area'].fillna(0,inplace=True)
```

Since Garage Cars and Garage Area null value is the same and shows no indication of a built garage, the null will be replaced with 0.

```python
train[(train['Garage Type'].notnull()) & (train['Garage Yr Blt'].isnull())]
```

```python
train.drop(train.index[1712],inplace=True)
```

Since the Garage Yr Built has an extra null from the Garage Type, I have decided to drop the observation.

```python
train[(train['Garage Type'].isnull()) & (train['Garage Yr Blt'].isnull())].shape
```

```python
train['Garage Type'].fillna('None',inplace=True)
train['Garage Yr Blt'].fillna(0,inplace=True)
```

Since the observations for Garage Type and Garage Yr Blt coincides with each other, we can deduce that there is no garage present for these observations and can fill them with 'None' and 0

```python
sns.distplot(train['Lot Frontage'],kde=False);
```

```python
train['Lot Frontage'].fillna(train['Lot Frontage'].mean(),inplace=True)
```

Since the values for lot frontage cannot be equal to 0 and does not logically equals to zero, I will replace the null values with the mean lot frontage.

```python
train.drop(columns=['Alley','Misc Feature'],inplace=True)
```

Since the null values in the columns are too big (over 90%) I have decided to drop these columns.

```python
#Checking data values for outliers.
train.describe()
```

```python
train[train['Garage Yr Blt'] == 2207]
```

```python
train['Garage Yr Blt'].replace(2207.0,2007.0,inplace=True)
```

```python
#Checking for any other discrepancy in garage built year.
train['Garage Yr Blt'].describe()
```

```python
train.to_csv(path_or_buf='../datasets/train_clean.csv',index=False)
```

Since the year built is 2007 and the year it was remodelled then sold was 2007, we can assume that the observation with the abnomaly in garage built to be the year it was remodelled which is 2007. This could be an input error during the data collection process.
