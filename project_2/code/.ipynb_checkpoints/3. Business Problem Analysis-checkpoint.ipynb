## Problem Statement

In this project, we were tasked to solve a business problem using the Ames Housing Data. This will be a independent analysis from the Kaggle Submission in Part 1.

**In this notebook, I am looking to understand how individuals can look to increase their prices of their houses by improving their houses.** 

Understanding that building a completely new house on the existing land is not cost effective and would require a large sum of money to invest in, I will zoom in to features or variables that a homeowner is able to manipulate or influence (i.e. via renovation or home improvement, etc) and would require a relatively lower budget (as compared to changing the structural integrity of the house) and ultimately how it can help predict the price for future sales.

As we can see from the table above, there are features that are strongly correlated to each other that may suggest multicollinearity. To avoid this, I will combine or create an interaction term between them.

```python
#Combining basement ordinal values together
train['Bsmt Qual_Cond'] = train['Bsmt Qual'] + train['Bsmt Cond']
train.drop(columns=['Bsmt Qual','Bsmt Cond'],inplace=True)
```

```python
#Creating pool interaction term
train['Pool Interaction'] = train['Pool Area'] * train['Pool QC']
train.drop(columns=['Pool Area','Pool QC'],inplace=True)
```

```python
#Creating fireplace interaction term
train['Fireplace Interaction'] = train['Fireplaces'] * train['Fireplace Qu']
train.drop(columns=['Fireplaces','Fireplace Qu'],inplace=True)
```

```python
#Creating basement type and finish interaction term
train['BsmtFin Type_SF1'] = train['BsmtFin Type 1'] * train['BsmtFin SF 1']
train.drop(columns=['BsmtFin Type 1','BsmtFin SF 1'],inplace=True)

train['BsmtFin Type_SF2'] = train['BsmtFin Type 2'] * train['BsmtFin SF 2']
train.drop(columns=['BsmtFin Type 2','BsmtFin SF 2'],inplace=True)
```

## Data Modeling & Prediction

```python
X = train.drop(columns='SalePrice')
y = train['SalePrice']
```

```python
Xtrain, Xtest, ytrain, ytest =  train_test_split(X,y,random_state=42)
```

```python
Xtrain_s = ss.fit_transform(Xtrain)
Xtest_s = ss.transform(Xtest)
```

### Linear Model

```python
lr = LinearRegression()
lr.fit(Xtrain,ytrain)
```

```python
print('Baseline Linear Regression')

#Train R2 score
lr_train_r2 = lr.score(Xtrain,ytrain)
print('Train R^2 Score: {}'.format(lr_train_r2.round(4)))

#Test R2 score
lr_test_r2 = lr.score(Xtest,ytest)
print('Test R^2 Score: {}'.format(lr_test_r2.round(4)))

#Cross Validated R2 Score
lr_cross_val_r2 = cross_val_score(lr, Xtrain, ytrain, cv=5).mean()
print('Cross Validated Train R^2 Score: {}'.format(lr_cross_val_r2.round(4)))
```

```python
lr_y_pred = lr.predict(Xtest_s)
```

```python
lr_res = lr_y_pred - ytest
```

```python
plt.figure(figsize=(14,10))
sns.regplot(ytest,lr_y_pred,marker="+").set(xlabel='Actual', ylabel='Predicted',title='Linear Predicted vs Actual Values');
```

```python
plt.figure(figsize=(14,10))
sns.regplot(ytest,lr_res,marker="+").set(xlabel='Actual', ylabel='Residual',title='Linear Residual vs Actual Values');
```

Although the scores of the R^2 is high across test and cross validation of the data, the simple plot aboves shows that the model is not a good fit due to the existence of negative values. As such, this model is not a good model.

### Lasso Model

```python
optimal_lasso = LassoCV(n_alphas=500, cv=10, verbose=1)
optimal_lasso.fit(Xtrain_s, ytrain)
```

```python
optimal_lasso.alpha_
```

```python
ls = Lasso(alpha=optimal_lasso.alpha_)
ls.fit(Xtrain_s, ytrain)
```

```python
print('Lasso Model')

#Train R2 score
ls_train_r2 = ls.score(Xtrain_s,ytrain)
print('Train R^2 Score: {}'.format(ls_train_r2.round(4)))

#Test R2 score
ls_test_r2 = ls.score(Xtest_s,ytest)
print('Test R^2 Score: {}'.format(ls_test_r2.round(4)))

#Cross Validated R2 Score
ls_cross_val_r2 = cross_val_score(ls, Xtrain_s, ytrain, cv=5).mean()
print('Cross Validated Train R^2 Score: {}'.format(ls_cross_val_r2.round(4)))
```

```python
ls_y_pred = ls.predict(Xtest_s)
ls_residuals = ytest - ls_y_pred
```

```python
plt.figure(figsize=(14,10))
sns.regplot(ytest,ls_y_pred,marker="+").set(xlabel='Actual', ylabel='Residual',title='Lasso Predicted vs Actual Values');
```

```python
plt.figure(figsize=(14,10))
sns.regplot(ytest,ls_residuals,marker="+").set(xlabel='Actual', ylabel='Residual',title='Lasso Residual vs Actual Values');
```

Unlike the baseline Linear Regression Model, the Lasso model generates predictions that are consistent with what we are looking for and produces a good R^2 score across test and cross validation data.

Let's now take a look at the MSE of the model.

### Ridge Model

```python
ridge_alphas = np.logspace(0, 5, 200)

optimal_ridge = RidgeCV(alphas=ridge_alphas, cv=10)
optimal_ridge.fit(Xtrain_s, ytrain)
optimal_ridge.alpha_
```

```python
ridge = Ridge(alpha=optimal_ridge.alpha_)
ridge.fit(Xtrain_s,ytrain)
```

```python
ridge_y_pred = ridge.predict(Xtest_s)
ridge_residuals = ytest - ridge_y_pred
```

```python
print('Ridge Model')

#Train R2 score
ridge_train_r2 = ridge.score(Xtrain_s,ytrain)
print('Train R^2 Score: {}'.format(ridge_train_r2.round(4)))

#Test R2 score
ridge_test_r2 = ridge.score(Xtest_s,ytest)
print('Test R^2 Score: {}'.format(ridge_test_r2.round(4)))

#Cross Validated R2 Score
ridge_cross_val_r2 = cross_val_score(ridge, Xtrain_s, ytrain, cv=5).mean()
print('Cross Validated Train R^2 Score: {}'.format(ridge_cross_val_r2.round(4)))
```

```python
plt.figure(figsize=(14,10))
sns.regplot(ytest,ridge_y_pred,marker="+").set(xlabel='Actual', ylabel='Predicted',title='Ridge Predicted vs Actual Values');
```

```python
plt.figure(figsize=(14,10))
sns.regplot(ytest,ridge_residuals,marker="+").set(xlabel='Actual', ylabel='Residual',title='Ridge Residual vs Actual Values');
```

Similar to the Lasso model, the Ridge Model generates predictions that are consistent with what we are looking for and produces a good R^2 score across test and cross validation data.

## Conclusion

**Findings**

- There were a large number of null values in the dataset. Instead of dropping them, we filled them methodically, always having a logical reason behind why we are filling a certain value in.

- From our data exploration and various modelling using lasso regression, we learnt that certain features such as (BsmtFin Type_SF1, Exter Qual and Kitchen Qual) were important to increase price of a home. Having such knowledge may help an owner to decide on what are the more important areas to improve on when trying to raise the value for future sales.

- We also saw that there was a strong multicollinearity between features. Systematically creating interaction terms and combining features have helped to deal with these multicollinearity. Although there were other highly correlated features still present, we did not combine them as there was no logical way to do so.

**Limitations**

- We have dropped features with a large value of null data that could not be logically filled. That may have affected our model.
- Strong multicollinearity between the feautures of the dataset may still exist, albeit us mitigating them and could still affect our model.

**Further Exploration**

- We have only worked with Linear, Lasso and Ridge models in this project. There may exist other models that could be a better fit that were not explored.
- Further feature engineering could also be done as we discover more on the relationships between features that may produce a better output for our model.
- Further comparative study on the prices of the works that could be done to the house would allow us to understand the cost and benefit of the changes with respect to the Sale Price of the house.
